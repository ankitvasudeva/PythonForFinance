```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
def getNSEMovingAvgPlot(symbol,column,interval, deviation):
    df = pd.read_csv('~/Notebook/data/stocks/nse/'+symbol+'.csv',parse_dates=True,index_col=0)
    df['MovingAvg'] = df[column].rolling(window=interval).mean()
    df['Std'] = df[column].rolling(window=interval).std()
    df['Upper'] = df['MovingAvg'] + df['Std']*deviation
    df['Lower'] = df['MovingAvg'] - df['Std']*deviation
    
    fig = go.Figure(data=[go.Candlestick(x=df.index,
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])
    fig.add_trace(go.Scatter(x=df.index, y=df['MovingAvg']))
    fig.add_trace(go.Scatter(x=df.index, y=df['Upper']))
    fig.add_trace(go.Scatter(x=df.index, y=df['Lower']))
    fig.show()
    df['%Upper_deviate'] = (df[column]-df['Upper'])/df['Std']
    df['%Avg_deviate'] = (df[column]-df['MovingAvg'])/df['Std']
    df['%Lower_deviate'] = (df[column]-df['Lower'])/df['Std']
    return df
```

```python
stockData= getNSEMovingAvgPlot('ITC_Yearly','Close',8,1)
```

```python
filterData = stockData.loc[((stockData['%Upper_deviate'] <= 0.1) & (stockData['%Upper_deviate'] >= -0.1)) | ((stockData['%Avg_deviate'] <= 0.1) & (stockData['%Avg_deviate'] >= -0.1))  | ((stockData['%Lower_deviate'] <= 0.1) & (stockData['%Lower_deviate'] >= -0.1))]
print(filterData)
```
