```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
def getYahooMovingAvgPlot(symbol,interval, deviation):
    df = pd.read_csv('~/Notebook/data/stocks/yf/'+symbol+'.csv',parse_dates=True,index_col=0)
    df['MovingAvg'] = df['Close'].rolling(window=interval).mean()
    df['Std'] = df['Close'].rolling(window=interval).std()
    df['Upper'] = df['MovingAvg'] + df['Std']*deviation
    df['Lower'] = df['MovingAvg'] - df['Std']*deviation
    fig = go.Figure(data=[go.Candlestick(x=df.index,
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])
    fig.add_trace(go.Scatter(x=df.index, y=df['MovingAvg']))
    fig.add_trace(go.Scatter(x=df.index, y=df['Upper']))
    fig.add_trace(go.Scatter(x=df.index, y=df['Lower']))
    fig.show()
    #checkDeviation(df)
    return df
```

```python
def checkDeviation(df):
    fields = ["Close", "High","Low"]
    for field in fields:
        df[field+'_Upper'] = (df[field]-df['Upper'])/df['Std']
        df[field+'_MovingAvg'] = (df[field]-df['MovingAvg'])/df['Std']
        df[field+'_Lower'] = (df[field]-df['Lower'])/df['Std']
    return df
```
