```python
import pandas as pd
import datetime as dt
from nsepy import get_history
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
style.use('ggplot')
```

```python
def getMovingAvgPlot(symbol,interval,column):
    df = pd.read_csv(symbol+'.csv',parse_dates=True,index_col=0)
    df['MovingAvg'] = df[column].rolling(window=interval).mean()
    ax1 = plt.subplot2grid((6,1), (0,0), rowspan=5, colspan=1)
    ax1.plot(df.index, df[column])
    ax1.plot(df.index, df['MovingAvg'])
    return
```

```python
getMovingAvgPlot('INFY',500,'Close')
```
